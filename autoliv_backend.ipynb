```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

def polynomial_regression(x, y, degree):
    x = np.array(x).reshape(-1, 1)
    
    # Transform the input data to include polynomial terms
    poly_features = PolynomialFeatures(degree=degree)
    x_poly = poly_features.fit_transform(x)
    
    # Fit the polynomial regression model
    model = LinearRegression()
    model.fit(x_poly, y)
    
    # Get the coefficients and intercept
    coefficients = model.coef_
    intercept = model.intercept_
    
    # Construct the polynomial function
    def polynomial_function(x):
        x = np.array(x).reshape(-1, 1)
        x_poly = poly_features.transform(x)
        return model.predict(x_poly)
    
    return polynomial_function, coefficients, intercept

# Example usage:
x = np.array(strain_interp1_new)  # Convert to numpy array
y = np.array(stress_avg_new)      # Convert to numpy array

# Add the (0, 0) point to the original data
x = np.insert(x, 0, 0)
y = np.insert(y, 0, 0)

# Find the index where strain crosses max
index_before = np.where(x <= x[index])[0][-1]
index_after = np.where(x >= x[index])[0][0]
index = index_before  # Choose index_before as the reference point

# Now use the data up to this index for polynomial fitting
x_part1 = x[:index + 1]
y_part1 = y[:index + 1]

degree = int(input("Enter the degree of polynomial:"))
poly_func, coeffs, intercept = polynomial_regression(x_part1, y_part1, degree)

print('Polynomial Coefficients:', coeffs)
print('Intercept:', intercept)

# Predict new values for the first part of the curve
y_new = poly_func(x_part1)

# Plotting
plt.figure(figsize=(10, 6))

# Plot the original data for the entire curve with small markers
plt.plot(x, y, 'bo-', label='Original Data (Full Curve)', markersize=0, linewidth=2)  # Adjust markersize here

# Overlay the polynomial fit for the first part of the curve
plt.plot(x_part1, y_new, 'r-', label='Polynomial Fit (Part 1)', linewidth=2)   # Red for polynomial fit

# Customize the plot
plt.title('Polynomial Fit for First Part of the Curve')
plt.xlabel('Strain')
plt.ylabel('Average Stress')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the power function
def power_function(x, a, b):
    return a * np.power(x, b)

# Example usage:
x = np.array(strain_interp1_new)  # Convert to numpy array
y = np.array(stress_avg_new)      # Convert to numpy array

# Add the (0, 0) point to the original data
x = np.insert(x, 0, 0)
y = np.insert(y, 0, 0)

# Find the index where strain crosses 10
index_before = np.where(x <= x[index])[0][-1]
index_after = np.where(x >= x[index])[0][0]
index = index_before  # Choose index_before as the reference point

# Now use the data up to this index for power function fitting
x_part1 = x[:index + 1]
y_part1 = y[:index + 1]

# Fit the power function to the data
params, covariance = curve_fit(power_function, x_part1, y_part1, p0=[1, 1])
y_new_power = power_function(x_part1, *params)

# Plotting
plt.figure(figsize=(10, 6))

# Plot the original data for the entire curve with small markers
plt.plot(x, y, 'bo-', label='Original Data (Full Curve)', markersize=0, linewidth=2)

# Overlay the power fit for the first part of the curve
plt.plot(x_part1, y_new_power, 'c-', label='Power Fit (Part 1)', linewidth=2)

# Customize the plot
plt.title('Power Fit for First Part of the Curve')
plt.xlabel('Strain')
plt.ylabel('Average Stress')
plt.legend()
plt.grid(True)
plt.show()
```
